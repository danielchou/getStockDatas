```python
import datetime
import requests
import pandas as pd
from io import StringIO
import pandas
pd.set_option('display.expand_frame_repr', False)  # 別折行

res = requests.get(
    "https://github.com/pandas-dev/pandas/blob/master/doc/data/titanic.csv")
dfs = pd.read_html(StringIO(res.text))
titanic = dfs[0]

```

### 如何依照年紀來排大小?

```python
titanic.sort_values(by="Age", ascending=False).head(12)
```

### 按照艙位、年紀由大排到小 

```python
#index會跟著跑
titanic.sort_values(by=['Pclass', 'Age'], ascending=False).head(10)  
```

```python
import pandas as pd
from finlab.data import Data
from sqlalchemy import create_engine, event
from talib import abstract
import _beowSet as bs          #自己寫的
from pandas.tseries.frequencies import to_offset

data = Data()
C     = data.get("收盤價").resample("W", offset="-5D").last()
ma4   = C.ewm(span=4, adjust=True).mean().round(2).stack("stock_id")
ma13  = C.ewm(span=13, adjust=True).mean().round(2).stack("stock_id")
ma52  = C.ewm(span=52, adjust=True).mean().round(2).stack("stock_id")
ma208 = C.ewm(span=208, adjust=True).mean().round(2).stack("stock_id")
H     = data.get("最高價").resample("W", offset="-5D").max().stack("stock_id")
L     = data.get("最低價").resample("W", offset="-5D").min().stack("stock_id")
O     = data.get("開盤價").resample("W", offset="-5D").first().stack("stock_id")

dfW = pd.DataFrame({"O": O, "H": H, "L": L, "C": C.stack("stock_id"), "ma4":ma4, "ma13": ma13,"ma52": ma52,"ma208": ma208})
dfV = pd.DataFrame({"vol": round(data.get("成交股數")/1000).resample("W").sum().stack("stock_id") })

dfW = dfW.merge(dfV, left_index=True, right_on=["date","stock_id"])
dfW.loc[(slice(None),"9962"),:]
```

```python
import pandas as pd
from finlab.data import Data
from sqlalchemy import create_engine, event
from talib import abstract
import _beowSet as bs          #自己寫的
from pandas.tseries.frequencies import to_offset

data = Data()
C     = data.get("收盤價").resample("W", offset="-5D").last()
ma4   = C.ewm(span=4, adjust=True).mean().round(2).stack("stock_id")
ma13  = C.ewm(span=13, adjust=True).mean().round(2).stack("stock_id")
ma52  = C.ewm(span=52, adjust=True).mean().round(2).stack("stock_id")
ma208 = C.ewm(span=208, adjust=True).mean().round(2).stack("stock_id")
H     = data.get("最高價").resample("W", offset="-5D").max().stack("stock_id")
L     = data.get("最低價").resample("W", offset="-5D").min().stack("stock_id")
O     = data.get("開盤價").resample("W", offset="-5D").first().stack("stock_id")

dfW = pd.DataFrame({"O": O, "H": H, "L": L, "C": C.stack("stock_id"), "ma4":ma4, "ma13": ma13,"ma52": ma52,"ma208": ma208})
dfV = pd.DataFrame({"vol": round(data.get("成交股數")/1000).resample("W").sum().stack("stock_id") })

dfW = dfW.merge(dfV, left_index=True, right_on=["date","stock_id"])
dfW.loc[(slice(None),"9962"),:]
```

```python

import numpy as np               # basic        ###### 
import pandas as pd              #   
import pandas_datareader as pdr  # get data     ######
import matplotlib.pyplot as plt  # visual       ######
import mplfinance as mpf        
%matplotlib inline
import seaborn as sns
import datetime as datetime      #time          ######    
import talib

start = datetime.datetime(2021,3,1)
dfD = pdr.DataReader('1524.TW', 'yahoo', start=start)
dfD.index.name = 'Date'
aggregation = {'Open':'first', 'High':'max','Low':'min','Close':'last','Volume':'sum'}

dfW = dfD.resample("1W").agg(aggregation)
dfM = dfD.resample("1M").agg(aggregation)


myStyle = mpf.make_mpf_style(
            marketcolors= mpf.make_marketcolors(up='r',down='g',edge='white',wick={'up':'red','down':'g'},volume='gray',ohlc='black')
            ,mavcolors=['#aaaeee','#FF00FF','#1f77b4','#ff7f0e'])
kwargs=dict(volume=True, type='candle', tight_layout=True, style=myStyle, figscale=1.25, warn_too_much_data=1500)

dates = ["2022-02-21","2022-05-24"]
datepairs = [(d1,d2) for d1,d2 in zip(dates,dates[1:])]

tlines=[
    dict(tlines=datepairs,tline_use='high',colors='g', tline_method="point-to-point",linestyle='dotted', linewidths = 0.5),
    dict(tlines=datepairs,tline_use='low',colors='b'),
    dict(tlines=datepairs,tline_use=['open','close'],colors='r')
]


print(datepairs)
mpf.plot(dfD, **kwargs, title="Daily", mav=(5,20,60,240), tlines = tlines)
mpf.plot(dfW, **kwargs, title="Weekly", mav=(4,13,52,208), tlines = tlines)
mpf.plot(dfM, **kwargs, title="Monthly", mav=(4,12,48,120), tlines = tlines)
```
